```python
# 재귀함수

def fac(n):
    if n == 0:
        return 1
    return n * fac(n - 1)

print(fac(5))

# 패킹 / 언패킹

# x, y = 1, 2
# z = 1, 2, 3

# a, *b = 1, 2, 3, 4
# print(a, b)

def mysum_(a, b, c):
    return a + b + c

num_list = [10, 20, 30]

rlt = mysum_(num_list[0], num_list[1], num_list[2])
rlts = mysum_(*num_list)

print(rlt)
print(rlts)

def test(*values):
    for value in values:
        print(value)

print(test(1))
print(test(1, 2))
print(test(1, 2, 3, 4))

def test(*args):
    return args
```
